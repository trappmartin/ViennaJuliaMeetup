### loading data set for demonstration

```julia
using DataFrames

# load dataset "iris" from package "datasets"...
data = dataset("datasets", "iris")
DataFrames.describe(data)
```

```julia
using Gadfly

# plot iris dataset
plot(data, x="SepalLength", y="SepalWidth", color="Species", Geom.point)
```

## compute basic statistics using StatsBase package

```julia
using StatsBase

# get array
X = convert(Array, data[:,1:4])

# compute moments
println("overall mean: ", mean(X))
println("variable wise mean: ", mean(X, 1))
println("variable wise variance: ", var(X, 1))

# normalize data (z-score)
X_z = zscore(X, 1)

println("variable wise mean & variance: ", mean_and_var(X_z, 1))
```

## computing clusterings on data using Clustering package

```julia
using Clustering
using Distances

# cluster using kmeans into 3 clusters
# note that X is assumed as follows: samples to be columns and rows to be variables
Clusters = kmeans(X', 3; maxiter=200)

D = pairwise(Euclidean(), X', X')
Clusters = dbscan(D, 0.5, 3)

# plot iris dataset
plot(data, x="SepalLength", y="SepalWidth", color=assignments(Clusters), Geom.point)
```

## using Distributions package to work with prob. distributions

```julia
using Distributions

# fit Gaussian distributions for each class
classes = unique(data[:Species])

dists = MvNormal[]

for c in classes
    
    d = convert(Array, data[data[:Species] .== c, [1:4]])
    m = mean(d, 1)
    cc = cov(d)

    push!(dists, MvNormal(vec(m), cc))
    
end

# make assignments according to likelihood
llh = reduce(hcat, map(dist -> logpdf(dist, convert(Array, data[[1:4]])'), dists) )

# get assignments
assignment = (findmax(llh', 1)[2] % 3)

# plot assignments
plot(data, x="SepalLength", y="SepalWidth", color=assignment, Geom.point)
```

## using MultivariateStats package for (mainly) dimension reduction

```julia
using MultivariateStats

# apply PCA on data
M = fit(PCA, X_z'; maxoutdim=2)
X_pca = transform(M, X_z')

# plot iris dataset
plot(data, x=X_pca[1,:], y=X_pca[2,:], color="Species", Geom.point)
```

## more state of the art Dimension reduction

```julia
#Pkg.clone("git://github.com/lejon/TSne.jl.git")

using TSne

Y = tsne(X_z, 2, 50, 1000, 20.0);

# plot iris dataset
plot(data, x=Y[:,1], y=Y[:,2], color="Species", Geom.point)
```

```julia
Clusters2 = kmeans(Y', 3; maxiter=200)

# plot iris dataset
p1 = plot(data, x="SepalLength", y="SepalWidth", color=assignments(Clusters2), Geom.point)
p2 = plot(data, x="SepalLength", y="SepalWidth", color="Species", Geom.point)

vstack(p1, p2)
```
